### Churn by Tenure Analysis

In this section, tenure was analysed to understand whether the length of time a customer stays with the company is related to churn.

The boxplot shows a clear difference between customers who stayed and customers who churned. Customers with `Churn = 0` generally have higher tenure, meaning they have stayed with the company for a longer period. In contrast, customers with `Churn = 1` are mostly concentrated in lower tenure values, showing that many churned customers left the company during the earlier months of their relationship.

The histogram confirms the same pattern. A higher number of churned customers can be seen in the lower tenure range, especially among newer customers. As tenure increases, the number of churned customers decreases, while the number of retained customers becomes stronger.

This suggests that customer churn is higher among newer customers. Therefore, the company should focus more on early-stage customer retention strategies, especially during the first months of service. Improving onboarding, offering early loyalty benefits, and monitoring new customer satisfaction could help reduce churn.

So lets try handling class imbalance becuase Some parameters not good 

```python
log_balanced = LogisticRegression(
    max_iter=1000,
    class_weight='balanced'
)
log_balanced.fit(x_train_scaler,y_train)
log_balanced_predict = log_balanced.predict(x_test_scaler)
```

```python
accuracy_score(y_test,log_balanced_predict)
```

```python
confusion_matrix(y_test,log_balanced_predict)
```

```python
print(classification_report(y_test,log_balanced_predict))
```

After applying class_weight='balanced', the logistic regression model improved significantly in detecting churn customers. The recall for the churn class increased from around 52% to 79%, meaning the model now identifies a much larger proportion of actual churn customers. However, this improvement comes with a decrease in overall accuracy and an increase in false positives. In a churn prediction context, this trade-off may be acceptable because missing actual churn customers is usually more costly than incorrectly flagging some non-churn customers.


Model Interpretation: Logistic Regression Coefficients

After selecting Logistic Regression with SMOTE as one of the strongest overall models, the model coefficients were analysed to understand which features had the strongest influence on customer churn. In Logistic Regression, a positive coefficient indicates that a feature increases the likelihood of churn, while a negative coefficient indicates that a feature reduces the likelihood of churn. The absolute coefficient value was also considered to identify the strongest predictors, regardless of whether their effect was positive or negative.

The results show that tenure is the most important factor in reducing churn. Its coefficient is strongly negative, meaning that customers who have stayed with the company for a longer period are less likely to leave. This supports the earlier exploratory analysis, where customers with shorter tenure had a much higher churn rate. Therefore, customer retention efforts should be focused especially on new customers during their first months of service.

The feature InternetService_Fiber optic has a strong positive coefficient, which means that customers using fiber optic internet are more likely to churn. This finding is also consistent with the EDA results, where fiber optic customers showed a higher churn rate compared with DSL and customers with no internet service. This may suggest that fiber optic customers have higher expectations regarding speed, reliability, and service quality. If those expectations are not met, they may be more likely to cancel their service.

Contract-related features also play an important role. Features such as Contract_Two year and Contract_Risk_Low Risk have negative coefficients, indicating that longer contracts reduce the probability of churn. This means that customers on two-year contracts are more stable and less likely to leave compared with customers on month-to-month contracts. Similarly, one-year contracts also reduce churn risk, although the effect is weaker than two-year contracts. This confirms that contract length is one of the strongest business factors affecting customer retention.

Some additional services also influence churn. For example, OnlineSecurity_Yes has a negative coefficient, suggesting that customers who subscribe to online security services are less likely to churn. This may indicate that customers who use more protective or value-added services are more engaged with the company. On the other hand, features such as StreamingTV_Yes, StreamingMovies_Yes, MultipleLines_Yes, and PaperlessBilling_Yes have positive coefficients, meaning they are associated with a higher probability of churn. These features may be linked to customers with higher monthly charges or higher expectations from the service.

Overall, the coefficient analysis shows that the most important factors reducing churn are longer tenure, longer contract duration, online security services, and having dependents. In contrast, the main factors increasing churn are fiber optic internet service, some streaming-related services, multiple lines, and paperless billing. These findings provide useful business insights because they show which customer groups are more at risk and where customer retention strategies should be focuse
